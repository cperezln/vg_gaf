```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from gaf.gaf import GramianDifferenceField, GramianSummationField
from utils.draw import draw_gaf, get_filtered, plot_relevant
```

```python
n = 30
seed = 100
timesteps = list(range(n))
periodic = np.sin(np.linspace(0, 4*np.pi, n))
np.random.seed(seed=seed)
gaussian_noise = np.random.randn(n)
combination = periodic + 0.3 * gaussian_noise
```

```python
ax1 = plt.subplot(2, 2, 1)
ax1.plot(periodic)
ax1.set_title("Periodic")
ax2 = plt.subplot(2, 2, 2)
ax2.plot(gaussian_noise)
ax2.set_title("Random noise")
ax3 = plt.subplot(2, 2, (3, 4))
ax3.plot(combination)
ax3.set_title("Periodic + Noise")
plt.subplots_adjust(hspace=0.4)
plt.show()
```

```python
gd11 = GramianDifferenceField(series = periodic, f=np.sin)
gd12 = GramianDifferenceField(series = periodic, f=np.cos)
gs11 = GramianSummationField(series= periodic, f=np.sin)
gs12 = GramianSummationField(series= periodic, f=np.cos)
```
