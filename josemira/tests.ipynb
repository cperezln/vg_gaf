```python
import numpy as np
n = 10
```

```python
thresholds = np.linspace(0, 3, num=n)
```

```python
%matplotlib inline
```

```python
print(period)
for idx, tau in enumerate(thresholds):
    gaf = GAF(xs, threshold=tau)
    gs = gaf.gasf()
    gd = gaf.gadf()
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.imshow(gs)
    ax1.set_title(f"GASF: threshold = {tau:.2}")
    ax2.imshow(gd)
    ax2.set_title(f"GADF: threshold = {tau:.2}")
    plt.savefig(f"outputs/img/threshold/{tau:.2}.png", dpi=100)
    plt.show()
```

```python
for idx, tau in enumerate(thresholds):
    gaf = GAF(xs, threshold=tau)
    gs = gaf.gasf()
    gd = gaf.gadf()
    gsD = abs(gs - gs0)
    gdD = abs(gd - gd0)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.imshow(gsD)
    ax1.set_title(f"GASF Difference: threshold = {tau:.2}")
    ax2.imshow(gdD)
    ax2.set_title(f"GADF Difference: threshold = {tau:.2}")
    plt.savefig(f"outputs/img/threshold/{tau:.2}-diff.png", dpi=100)
    plt.show()
```
